```python
fs = int(eeg["srate"])
noise_blocks = [eeg["noise"][i] for i in range(eeg["noise"].shape[0])]  # list of (channels, samples)

def noise_psd_summary(block, fs, nperseg=None, overlap=0.5, prominence=0.01):
    """
    Returns freq grid, channel-avg PSD, and a small table of top peaks.
    block: (channels, samples)
    """
    x = np.asarray(block, dtype=float)
    if nperseg is None:
        nperseg = min(fs*2, x.shape[1])  # 2 s window, or shorter if needed
    noverlap = int(nperseg*overlap)
    # PSD per channel, then average across channels
    f, Pxx = welch(x, fs=fs, nperseg=nperseg, noverlap=noverlap, axis=1)  # Pxx shape: (channels, freq)
    Pxx_mean = Pxx.mean(axis=0)

    # find peaks on the averaged PSD (log-space helps; keep it simple here)
    peaks, props = find_peaks(Pxx_mean, prominence=prominence * np.max(Pxx_mean))
    peak_freqs = f[peaks]
    peak_vals  = Pxx_mean[peaks]
    # sort peaks by magnitude (desc)
    order = np.argsort(peak_vals)[::-1]
    peak_freqs = peak_freqs[order]
    peak_vals  = peak_vals[order]

    return f, Pxx_mean, peak_freqs, peak_vals

# Compute per-block PSD + peaks
psd_results = [noise_psd_summary(nb, fs, nperseg=fs*2, overlap=0.5, prominence=0.02) for nb in noise_blocks]

# Plot them overlaid
plt.figure(figsize=(10,5))
for i, (f, P, pf, pv) in enumerate(psd_results):
    plt.semilogy(f, P, label=f"Noise block {i+1}")
plt.axvline(60, color='k', linestyle='--', alpha=0.5, label='60 Hz (US mains)')
plt.axvline(120, color='k', linestyle=':',  alpha=0.4, label='120 Hz')
plt.axvline(180, color='k', linestyle=':',  alpha=0.3, label='180 Hz')
plt.xlabel("Frequency (Hz)")
plt.ylabel("Power (V/Hz)")
plt.title("Noise PSD (channel-averaged)")
plt.grid(True, which='both', ls='--', alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

# Print top 5 peaks per block
for i, (f, P, pf, pv) in enumerate(psd_results):
    print(f"\nTop peaks  Noise block {i+1}:")
    for k in range(min(5, len(pf))):
        print(f"  {pf[k]:6.1f} Hz  | PSD={pv[k]:.3g}")
```

```python
clean_noise = apply_basic_filters(noise_blocks[0], fs, hp=1.0, lp=40.0, notch_base=60.0)
f, P, pf, pv = noise_psd_summary(clean_noise, fs)
plt.semilogy(f, P)
plt.title("Noise after HP+LP+Notch filters")
plt.xlabel("Frequency (Hz)"); plt.ylabel("Power (V/Hz)")
plt.grid(True, which='both', ls='--', alpha=0.3)
plt.show()
```
