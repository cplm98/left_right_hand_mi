```python
def print_all_keys_and_nested_keys(data):
    """
    Prints all keys in a dictionary and recursively prints keys of nested dictionaries.

    Args:
        data (dict): The dictionary to process.
    """
    for key, value in data.items():
        print(key)
        if isinstance(value, dict):
            print_all_keys_and_nested_keys(value)
```

```python
from scipy.signal import iirnotch, filtfilt, butter
```

```python
def notch_filter(data, fs, freq=60.0, q=30.0):
    """
    Apply a single 60 Hz (default) IIR notch to data with sampling rate fs.
    data: np.ndarray of shape (..., samples) or (samples,)
    fs: sampling rate (e.g., 512)
    freq: notch center frequency in Hz (e.g., 60.0)
    q: quality factor; higher = narrower notch (typical 2050)
    """
    b, a = iirnotch(w0=freq, Q=q, fs=fs)
    # Apply along the last axis (time)
    return filtfilt(b, a, data, axis=-1)

def notch_filter_harmonics(data, fs, base=60.0, n_harmonics=3, q=30.0, nyquist_guard=0.9):
    """
    Apply notches at base, 2*base, 3*base, ... up to n_harmonics,
    skipping any harmonic that would exceed a guard fraction of Nyquist.
    """
    out = np.array(data, copy=True)
    nyq = fs / 2
    for k in range(1, n_harmonics+1):
        f0 = k * base
        if f0 < nyquist_guard * nyq:
            b, a = iirnotch(w0=f0, Q=q, fs=fs)
            out = filtfilt(b, a, out, axis=-1)
    return out
```

```python
# ---------- Core Butterworth filters (zero-phase) ----------

def butter_filter(data, fs, hp=None, lp=None, order=4):
    """
    Zero-phase Butterworth filtering with optional high- and/or low-pass.
    data: array-like (..., samples)
    fs: sampling rate (Hz)
    hp: high-pass cutoff in Hz (None to skip)
    lp: low-pass cutoff in Hz (None to skip)
    order: filter order (per section)
    """
    x = np.asarray(data, dtype=float, copy=True)
    nyq = fs / 2.0
    if hp is not None:
        if not (0 < hp < nyq):
            raise ValueError(f"hp must be in (0, {nyq})")
        b, a = butter(order, hp/nyq, btype='highpass')
        x = filtfilt(b, a, x, axis=-1)
    if lp is not None:
        if not (0 < lp < nyq):
            raise ValueError(f"lp must be in (0, {nyq})")
        b, a = butter(order, lp/nyq, btype='lowpass')
        x = filtfilt(b, a, x, axis=-1)
    return x

def butter_bandpass(data, fs, f_lo, f_hi, order=4):
    """Zero-phase Butterworth band-pass (f_lo < f_hi)."""
    if not (0 < f_lo < f_hi < fs/2):
        raise ValueError("Cutoffs must satisfy 0 < f_lo < f_hi < fs/2")
    x = np.asarray(data, dtype=float, copy=True)
    nyq = fs / 2.0
    b, a = butter(order, [f_lo/nyq, f_hi/nyq], btype='bandpass')
    return filtfilt(b, a, x, axis=-1)

# ---------- Notch @ 60 Hz (and harmonics) ----------

def notch_60hz(data, fs, q=30.0):
    """
    Single-frequency IIR notch at 60 Hz (US mains), zero-phase.
    q: quality factor (higher => narrower notch). Typical 2050.
    """
    b, a = iirnotch(w0=60.0, Q=q, fs=fs)
    return filtfilt(b, a, np.asarray(data, dtype=float, copy=True), axis=-1)

def notch_harmonics(data, fs, base=60.0, n_harmonics=3, q=30.0, nyquist_guard=0.9):
    """
    Apply notches at base, 2*base, ..., up to n_harmonics.
    Skips any harmonic >= nyquist_guard * (fs/2) to avoid instability.
    """
    x = np.asarray(data, dtype=float, copy=True)
    nyq = fs / 2.0
    for k in range(1, n_harmonics+1):
        f0 = k * base
        if f0 < nyquist_guard * nyq:
            b, a = iirnotch(w0=f0, Q=q, fs=fs)
            x = filtfilt(b, a, x, axis=-1)
    return x

# ---------- Convenience: typical EEG cleaning sequence ----------

def apply_basic_filters(data, fs, hp=1.0, lp=40.0, notch_base=60.0, n_harm=2, q=30.0, order=4):
    """
    Typical order: High-pass -> Notch(es) -> Low-pass (all zero-phase).
    Returns filtered array with the same shape as input.
    """
    x = butter_filter(data, fs, hp=hp, lp=None, order=order)
    x = notch_harmonics(x, fs, base=notch_base, n_harmonics=n_harm, q=q)
    x = butter_filter(x, fs, hp=None, lp=lp, order=order)
    return x
```
